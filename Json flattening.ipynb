```python
import os
os.chdir(r'C:\Users\nitin\Documents\MSBA_6420_Predictive Analytics\Team_Project')
import json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

# Loading the train data in chunks

```python
file = 'train_v2.csv'
JSON_COLUMNS = ['device', 'geoNetwork', 'totals', 'trafficSource']
df_chunk = pd.read_csv(file, chunksize=1000000,
                       dtype={'fullVisitorId': 'str'})
```

```python
df_chunk
```

# Concating the chunks into single Dataframe

```python
chunk_list = []  

for chunk in df_chunk: 
    chunk_list.append(chunk)
    
train = pd.concat(chunk_list)
```

# Removing the unwanted columns

```python
train = train.drop(['customDimensions','hits'], axis = 1)
```

```python
train.shape
```

```python
train.head()
```

# Exporting the train data(reduced size) to csv

```python
train.to_csv(r'C:\Users\nitin\Documents\MSBA_6420_Predictive Analytics\Team_Project\train_json.csv')
```

# Flattening of Json columns in train data

```python
df = pd.read_csv('train_json.csv', 
                     converters={column: json.loads for column in JSON_COLUMNS}, 
                     dtype={'fullVisitorId': 'str'}, # Important!!
                     )

for column in JSON_COLUMNS:
        column_as_df = json_normalize(df[column])
        column_as_df.columns = [f"{column}.{subcolumn}" for subcolumn in column_as_df.columns]
        df = df.drop(column, axis=1).merge(column_as_df, right_index=True, left_index=True)
```
